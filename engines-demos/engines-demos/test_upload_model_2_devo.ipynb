```python
#define variables
MODEL_SERVER = 'https://api.stage.devo.com/makeijan' #fill with makeijan endpoint
DOMAIN = 'self' #fill with the user domain
DEVO_TOKEN =  '262b0423715de3b5f5a63303eb2a96ba'  #fill with user token
```

```python
#initialize the model server
from makeijan_client.custom import ModelServer
#from makeijan_client.custom.model_server import ModelServer
from makeijan_client.rest import ApiException

model_server = ModelServer(MODEL_SERVER, DOMAIN, DEVO_TOKEN)
model_server._connect()
```

```python
#model variables
MODEL_PATH = '/content/drive/MyDrive/Colab Notebooks/engines-demos/artifacts/h2o/autoencoder_h2o.zip' #fill with the path to the file with the model ex. /artifacts/autoencoder_h2o.zip
MODEL_NAME = 'credit_card' #fill with model name to be used by the query engine ex. credit_card
MODEL_DESC = 'testing from google collab ' #fill with some description about model functionality
MODEL_ENGINE = 'H2O' #fill with the engine used to build the model
```

```python
#insert/update the model
try:
  with model_server:
    res = model_server.find_model(MODEL_NAME)
    if (res):
      model_server.update_image(MODEL_NAME, 
                                MODEL_PATH)
      print("The model has been updated.")
    else:
      model_server.create_model(MODEL_NAME,
                                MODEL_ENGINE,
                                MODEL_PATH,
                                MODEL_DESC)
      print("The model has been created.")

except ApiException as e:
  print('There was a problem inserting the model.')
  raise e
```
